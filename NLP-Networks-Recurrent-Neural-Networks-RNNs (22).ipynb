```python
# importing required library

import torch
import torch.nn as nn

# Defining a simple RNN model
class SimpleRNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(SimpleRNN, self).__init__()
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
    
    def forward(self, x):
        out, _ = self.rnn(x)
        return out

# Instantiating the model
input_size = 10  # Replacing with required input size
hidden_size = 20  # Replacing with required  hidden size
num_layers = 2  # Replacing with the number of layers
model = SimpleRNN(input_size, hidden_size, num_layers)

# Defining toy sequence data and target
sequence = torch.randn(5, 3, input_size)  # Replacing 5 with the batch size
target = torch.randn(5, 3, hidden_size)  # Use hidden_size for the target

# Loss and optimization function
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Training loop for 100 epochs
num_epochs = 100
for epoch in range(num_epochs):
    optimizer.zero_grad()
    output = model(sequence)
    loss = criterion(output, target)
    loss.backward()
    optimizer.step()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```
