# Working with Bag of words

```python
# Bag of words is just counting the word frequency

# and research has proven that during sentiment analysis, bag of word is the best technique /

import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["people watch dsrajnish_adhikari",
                         "dsrajnish_adhikari watch dsrajnish_adhikari",
                         "people write comment",
                          "dsrajnish_adhikari write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#vocabulary
print(cv.vocabulary_)
```

```python
bow.toarray()
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
# new
cv.transform(['rajnish watch dsrajnish_adhikari']).toarray()
```

```python
X = bow.toarray()
y = df['output']
```

#  N-grams

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
#Ti gram
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

# TF-IDF (Term frequency- Inverse document frequency)

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid= TfidfVectorizer()
```

```python
tfid.fit_transform(df['text']).toarray()
```

```python
print(tfid.idf_)
```
